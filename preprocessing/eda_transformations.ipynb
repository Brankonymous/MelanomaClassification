```python
from constants import *
import csv
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df = pd.read_csv('../' + TRAIN_CSV_NAME)

idx = 5
image_path = '../' + TRAIN_DATASET_PATH + '/' + df.iloc[idx, 0] + '.jpg'
image = Image.open(image_path).convert('RGB')

def rotate_image(image, angle):
    rotated = image.rotate(angle)
    return rotated

rotated_image = rotate_image(image, 130) 
horizontal_flip_image = image.transpose(Image.FLIP_LEFT_RIGHT)
vertical_flip_image = image.transpose(Image.FLIP_TOP_BOTTOM)

# Plot the images
plt.figure(figsize=(10, 10))

# Original image
plt.subplot(2, 2, 1)
plt.imshow(image)
plt.title('Originalna slika')
plt.axis('off')

# Rotated 
plt.subplot(2, 2, 2)
plt.imshow(rotated_image)
plt.title('Rotirana slika (130 stepeni)')
plt.axis('off')

# Horizontally flipped
plt.subplot(2, 2, 3)
plt.imshow(horizontal_flip_image)
plt.title('Horizontalno-obrnuta slika')
plt.axis('off')

# Vertically flipped 
plt.subplot(2, 2, 4)
plt.imshow(vertical_flip_image)
plt.title('Vertikalno-obrnuta slika')
plt.axis('off')

plt.savefig('pictures/augmented_images.png')
plt.show()
```
